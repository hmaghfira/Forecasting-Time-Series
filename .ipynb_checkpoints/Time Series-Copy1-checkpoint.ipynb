# DATA UNDERSTANDING

### IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
import datetime
# import seaborn as sns
rcParams['figure.figsize'] = 10,6
```

### LOAD DATASET

```python
data = pd.read_csv('time.csv')

# rename Unnamed: 0 column
df=data.rename(columns={'Unnamed: 0':'Month'})
df.head()
```

```python
# change format to m-Y
date='%b-%y'
month=datetime.datetime.strptime('Mar-81', date)
month.strftime('%m-%y')
```

```python
# convert string date to datetime
df['Month']=df['Month'].apply(lambda x: datetime.datetime.strptime(x, date))
df.head()
```

```python
# get year only
df=df.set_index(['Month'])
# df['Year']=df['Month'].dt.year

df.head()
```

# DATA EXPLORATION & PREPARATION

##### MISSING VALUES

```python
#count how many missing values
count_missing = df.isnull().sum().sort_values(ascending = False)
count_missing.head(10)
```

##### CHOOSE THE USED COLUMNS

```python
# choose Year and Sales to be predicted by time series analysys algorithm
fin=df[['Sales']]
fin.head()
```

##### PLOT TO SEE THE STATIONARY

```python
## plot graph
plt.xlabel("Data")
plt.ylabel("Number of Sales")
plt.plot(fin)
```

Dari grafik di atas, dapat diketahui bahwa Hipotesis awal **(H0) tidak stasioner** pada ragam.

##### PROVING THE HIPOTHESIS

Whether it's already stationaire or not

**USING GRAPH**

```python
# Determing rolling statistics
rolmean = fin.rolling(window=4).mean()
rolstd = fin.rolling(window=4).std()
print(rolmean, rolstd)
```

```python
#Plot rolling statistics:
orig = plt.plot(fin, color='blue',label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

Dari trendline rolling std (representasi dari **keberagaman**), terlihat bahwa line menunjukkan garis yang cenderung **stasioner**. Sedangkan line rolling mean (representasi dari rataan) masih belum menunjukkan garis trendline yang jelas.

**USING DICKEY-FULLER**

```python
#Perform Dickey-Fuller test:
from statsmodels.tsa.stattools import adfuller

print ('Results of Dickey-Fuller Test:')
dftest = adfuller(fin['Sales'], autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
  dfoutput['Critical Value (%s)' %key] = value
  
print(dfoutput)
```

Terlihat bahwa p-value menunjukkan nilai yang kurang dari a (0,05), artinya menolak H0. Sehingga data **sudah stasioner terhadap ragam**.<br>
Selanjutnya akan dilakukan differencing untuk membuat stasioner terhadap rataan.

**DIFFERENCING USING LOG**

```python
#Estimating trend
indexedDataset_logScale = np.log(fin)
plt.plot(indexedDataset_logScale)
```

```python
# See the moving average
movingAverage = indexedDataset_logScale.rolling(window=4).mean()
movingSTD = indexedDataset_logScale.rolling(window=4).std()
plt.plot(indexedDataset_logScale)
plt.plot(movingAverage, color='red')
plt.plot(movingSTD, color='black')
```

```python
# Compiling differencing
datasetLogScaleMinusMovingAverage = indexedDataset_logScale - movingAverage
datasetLogScaleMinusMovingAverage.head(12)

# Remove Nan Values
datasetLogScaleMinusMovingAverage.dropna(inplace=True)
datasetLogScaleMinusMovingAverage.head(10)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
  
  #Determing rolling statistics
  movingAverage = timeseries.rolling(window=4).mean()
  movingSTD = timeseries.rolling(window=4).std()
  
  #Plot rolling statistics:
  orig = plt.plot(timeseries, color='blue',label='Original')
  mean = plt.plot(movingAverage, color='red', label='Rolling Mean')
  std = plt.plot(movingSTD, color='black', label='Rolling Std')
  plt.legend(loc='best')
  plt.title('Rolling Mean & Standard Deviation')
  plt.show(block=False)
  
  #Perform Dickey-Fuller test:
  print('Result of Dickey-Fuller Test:')
  dftest = adfuller(timeseries['Sales'], autolag='AIC')
  dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lag Used','Number of Observations Used'])
  for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
  print(dfoutput)
```

```python
# See the moving average
movingAverage = datasetLogScaleMinusMovingAverage.rolling(window=4).mean()
movingSTD = datasetLogScaleMinusMovingAverage.rolling(window=4).std()
plt.plot(datasetLogScaleMinusMovingAverage)
plt.plot(movingAverage, color='red')
plt.plot(movingSTD, color='black')
```

```python
test_stationarity(datasetLogScaleMinusMovingAverage)
```

```python
# do log scale again to see the mean
exponentialDecayWeightedAverage = indexedDataset_logScale.ewm(halflife=12, min_periods=0, adjust=True).mean()
plt.plot(indexedDataset_logScale)
plt.plot(exponentialDecayWeightedAverage, color='red')
```

```python
datasetLogScaleMinusMovingExponentialDecayAverage = indexedDataset_logScale -  exponentialDecayWeightedAverage
test_stationarity(datasetLogScaleMinusMovingExponentialDecayAverage)
```

```python
datasetLogDiffShifting = indexedDataset_logScale - indexedDataset_logScale.shift()
plt.plot(datasetLogDiffShifting)
```

```python
datasetLogDiffShifting.dropna(inplace=True)
test_stationarity(datasetLogDiffShifting)
```

Setelah dilakukan differencing sebanyak 3 kali, dapat dilihat bahwa nilai p-value semakin rendah menunjukkan bahwa data **semakin stasioner**.

##### SEE THE CHARACTERISTIC OF THE DATA

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(datasetLogScaleMinusMovingExponentialDecayAverage)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(datasetLogScaleMinusMovingExponentialDecayAverage, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()

decomposedLogData = residual
decomposedLogData.dropna(inplace=True)
test_stationarity(decomposedLogData)
```

Dari grafik di atas, dapat diketahui bahwa dataset memiliki karakteristik yang seasonal. Selanjutnya akan dilihat nilai PACF dan ACF untuk menentukan model yang tepat.

Akan menggunakan algoritma ARIMA.

```python
# find the best arima (less mse)

import warnings
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt

# evaluate an ARIMA model for a given order (p,d,q) and return RMSE
def evaluate_arima_model(X, arima_order):
# prepare training dataset
    X = X.astype('float32')
    train_size = int(len(X) * 0.50)
    train, test = X[0:train_size], X[train_size:]
    history = [x for x in train]
    # make predictions
    predictions = list()
    for t in range(len(test)):
        model = ARIMA(history, order=arima_order)
        model_fit = model.fit(disp=0)
        yhat = model_fit.forecast()[0]
        predictions.append(yhat)
        history.append(test[t])
    # calculate out of sample error
    mse = mean_squared_error(test, predictions)
    rmse = sqrt(mse)
    return rmse

# evaluate combinations of p, d and q values for an ARIMA model
def evaluate_models(dataset, p_values, d_values, q_values):
    dataset = dataset.astype('float32')
    best_score, best_cfg = float("inf"), None
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p,d,q)
                try:
                    mse = evaluate_arima_model(dataset, order)
                    if mse < best_score:
                        best_score, best_cfg = mse, order
                    print('ARIMA%s MSE=%.3f' % (order,mse))
                except:
                    continue
    print('Best ARIMA%s MSE=%.3f' % (best_cfg, best_score))

# load dataset
# series = pd.read_csv('time.csv')
# evaluate parameters
p_values = range(0,3)
d_values = range(0,3)
q_values = range(0,3)
warnings.filterwarnings("ignore")
evaluate_models(indexedDataset_logScale.values, p_values, d_values, q_values)
```
