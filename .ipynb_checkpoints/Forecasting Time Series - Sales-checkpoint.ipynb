```python
## plot graph
plt.xlabel("Year")
plt.ylabel("Number of Sales")
plt.plot(fin)
```

Dari grafik di atas, dapat diketahui bahwa Hipotesis awal **(H0) tidak stasioner** pada ragam.

### PROVING THE HIPOTHESIS

Whether it's already stationaire or not

**USING GRAPH**

```python
# Determing rolling statistics
rolmean = fin.rolling(window=4).mean() #window 4 dipilih karena tiap tahunnya memiliki 4 records
rolstd = fin.rolling(window=4).std()
print(rolmean, rolstd)
```

```python
# Buat ngeliat pola garis dari variansi dan mean apakah stasioner atau tidak
#Plot rolling statistics:
orig = plt.plot(fin, color='blue',label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

Dari trendline rolling std (representasi dari **keberagaman**), terlihat bahwa line menunjukkan garis yang cenderung **belum stasioner**. Sedangkan line rolling mean (representasi dari rataan) masih belum menunjukkan garis trendline yang jelas.

**USING DICKEY-FULLER**

```python
#Perform Dickey-Fuller test:
from statsmodels.tsa.stattools import adfuller
from pandas.core import datetools

print ('Results of Dickey-Fuller Test:')
dftest = adfuller(fin['Sales'], autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
  dfoutput['Critical Value (%s)' %key] = value
  
print(dfoutput)
```

Terlihat bahwa **p-value** menunjukkan nilai yang **kurang dari a (0,05)**, artinya menolak H0. Sehingga dapat dikatakan bahwa data **sudah stasioner terhadap ragam**.<br>
Selanjutnya akan dilakukan log untuk membuat stasioner terhadap rataan.

**USING LOG** TO HANDLE MEAN

```python
# Plot grafik
#Estimating trend using log
test_log = np.log(fin)
plt.plot(test_log)
```

```python
# See the moving average
movingAverage = test_log.rolling(window=4).mean()
movingSTD = test_log.rolling(window=4).std()
plt.plot(test_log)
plt.plot(movingAverage, color='red', label='movingAverage')
plt.plot(movingSTD, color='black', label='movingSTD')
plt.legend(loc='best')
plt.show(block=False)
```

```python
# Compiling differencing
dt_logMoving = test_log - movingAverage
dt_logMoving.head(12)

# Remove Nan Values
dt_logMoving.dropna(inplace=True)
dt_logMoving.head(10)
```

```python
# Melihat hasil yang udah di log
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
  
  #Determing rolling statistics
  movingAverage = timeseries.rolling(window=4).mean()
  movingSTD = timeseries.rolling(window=4).std()
  
  #Plot rolling statistics:
  orig = plt.plot(timeseries, color='blue',label='Original')
  mean = plt.plot(movingAverage, color='red', label='Rolling Mean')
  std = plt.plot(movingSTD, color='black', label='Rolling Std')
  plt.legend(loc='best')
  plt.title('Rolling Mean & Standard Deviation')
  plt.show(block=False)
  
  #Perform Dickey-Fuller test:
  print('Result of Dickey-Fuller Test:')
  dftest = adfuller(timeseries['Sales'], autolag='AIC')
  dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lag Used','Number of Observations Used'])
  for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
  print(dfoutput)
```

```python
test_stationarity(dt_logMoving)
```

```python
# see the line of exponential mean
exponentialDecayWeightedAverage = test_log.ewm(halflife=12, min_periods=0, adjust=True).mean()
plt.plot(test_log)
plt.plot(exponentialDecayWeightedAverage, color='red')
```

```python
dt_logMovingExp = test_log -  exponentialDecayWeightedAverage
test_stationarity(dt_logMovingExp)
```

Hasil dari log di atas terlihat bahwa garis mean dan std masih belum stasioner, oleh karena itu selanjutnya akan dilakukan differencing.

**USING DIFFERENCING** AFTER LOG

```python
# Differencing after log
test_diff = test_log - test_log.shift()
plt.plot(test_diff)
```

```python
# drop if there exists missing vallues
test_diff.dropna(inplace=True)
test_stationarity(test_diff)
```

Setelah dilakukan differencing, dapat dilihat bahwa nilai p-value semakin rendah di bawah nilai a (0,05). Hal ini menunjukkan bahwa data **semakin stasioner**. Selanjutnya akan dilihat karakteristik dari data.

### SEE THE CHARACTERISTIC OF THE DATA

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(dt_logMovingExp)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(dt_logMovingExp, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()

decomposedLogData = residual
decomposedLogData.dropna(inplace=True)
test_stationarity(decomposedLogData)
```

```python
# melihat nilai residual
decomposedLogData = residual
decomposedLogData.dropna(inplace=True)
test_stationarity(decomposedLogData)
```

Dari grafik di atas, dapat diketahui bahwa dataset memiliki karakteristik yang **seasonal**. Selanjutnya akan dilihat nilai PACF dan ACF untuk menentukan model yang tepat.

### SARIMAX MODEL

```python
#import warnings and itertools
import warnings
import itertools

p = d = q = range(0, 4) # Define the p, d and q parameters to take any value between 0 and 2
pdq = list(itertools.product(p, d, q)) # Generate all different combinations of p, q and q triplets
pdq_x_QDQs = [(x[0], x[1], x[2], 6) for x in list(itertools.product(p, d, q))] # Generate all different combinations of seasonal p, q and q triplets
print('Examples of Seasonal ARIMA parameter combinations for Seasonal ARIMA...')
print('Seasonal ARIMA: {} x {}'.format(pdq[1], pdq_x_QDQs[1]))
print('Seasonal ARIMA: {} x {}'.format(pdq[2], pdq_x_QDQs[2]))
print('Seasonal ARIMA: {} x {}'.format(pdq[3], pdq_x_QDQs[3]))
print('Seasonal ARIMA: {} x {}'.format(pdq[4], pdq_x_QDQs[4]))
```

```python
model = sm.tsa.statespace.SARIMAX(fin, order=(0,1,1), seasonal_order=(1,1,1,6), enforce_stationarity=False,enforce_invertibility=False)
results_model = model.fit()
results_model.summary()
```

```python
results_model.resid.plot()
```

```python
start_index = test.index.min()
end_index = test.index.max()

#Predictions
pred = results_model.get_prediction(start=start_index,end=end_index, dynamic=False)
```

```python
pred_ci = pred.conf_int()
ax = fin['1970':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Sales')
plt.legend()
plt.show()
```

```python
results_model.plot_diagnostics(figsize=(15, 12))
plt.show()
```

Dari grafik diagnostik di atas, dapat diketahui bahwa residual model terdistribusi secara normal :
1. Pada grafik kanan atas, garis merah KDE mendekati garis N(0,1), dimana N(0,1) adalah standar notasi dari distribusi normal dengan rata-rata 0 dan standar deviasi 1.
2. Pada grafik kiri bawah, QQ-plot menunjukkan bahwa distribusi residual (titik biru) mengikuti garis tren dari standar distribusi normal.
3. Pada grafik kiri atas, grafik residual tidak menunjukkan adanya pola seasonal. Hal ini diperkuat dengan grafik kanan bawah (autocorrelation) yang menunjukkan bahwa time series dari residual memiliki korelasi yang rendah dengan versi lagnya.
